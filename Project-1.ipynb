```python
import pandas as pd
import numpy as np

# Assuming Pre_int is your DataFrame
# List of years to analyze
years = [2009, 2010, 2011, 2012, 2013, 2014]  # Adjust this list as needed

# Initialize a list to hold the results for each year
results = []

for year in years:
    # Criteria for Camden City schools based on 'School Name'
    camden_city_criteria = ((Pre_int['School Name'].str.contains('CAMDEN CITY', case=False, na=False)) &
                            (Pre_int['YEAR'] == year))

    # Criteria for Charter schools based on 'District Name'
    charter_criteria = ((Pre_int['District Name'].str.contains('CHARTER', case=False, na=False)) &
                        (Pre_int['YEAR'] == year))

    # Combine both criteria
    combined_criteria = camden_city_criteria | charter_criteria

    # Filter the DataFrame based on the combined criteria
    df_filtered = Pre_int[combined_criteria]

    # Assuming the score and count column names
    score_columns = ['PP', 'P', 'AP', 'Mean Scale']
    count_column = 'Valid Scale'  # Column with the count of valid test takers

    # Calculate weighted sums, total test takers, and enrollment for the current year
    weighted_sums = (df_filtered[score_columns].T * df_filtered[count_column]).T.sum()
    total_valid_test_takers = df_filtered[count_column].sum()
    total_enrollment = df_filtered['Enroll'].sum()

    # Calculate weighted averages for score categories
    weighted_averages = weighted_sums / total_valid_test_takers

    # Append the results for the current year to the results list
    results.append({
        'Year': year,
        'PP': weighted_averages.get('PP', np.nan),
        'P': weighted_averages.get('P', np.nan),
        'AP': weighted_averages.get('AP', np.nan),
        'Mean Scale': weighted_averages.get('Mean Scale', np.nan),
        'Total Valid Test Takers': total_valid_test_takers,
        'Total Enrollment': total_enrollment
    })

# Convert the results list into a DataFrame
results_df = pd.DataFrame(results)

# Display the results DataFrame
print(results_df)
```
